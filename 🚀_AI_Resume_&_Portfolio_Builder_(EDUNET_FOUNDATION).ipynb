```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from reportlab.lib.pagesizes import A4
from reportlab.pdfgen import canvas
```

CELL 2: USER INPUT (SIMULATED)

```python
name = "Deepak A"
resume_text = """
Python Machine Learning HTML CSS Flask
Developed AI projects and web applications
"""
```

```python
job_roles = {
    "Web Developer": "HTML CSS JavaScript Python Flask",
    "AI Engineer": "Python Machine Learning NLP Deep Learning",
    "Data Analyst": "Python SQL Statistics Data Analysis"
}
```

CELL 3: ATS SCORE CALCULATION (AI MATCHING)

```python
vectorizer = TfidfVectorizer()

print("ATS Match Scores:\n")
role_scores = {}

for role, desc in job_roles.items():
    vectors = vectorizer.fit_transform([resume_text, desc])
    score = cosine_similarity(vectors[0], vectors[1])[0][0]
    role_scores[role] = score
    print(role, ":", round(score*100, 2), "%")
```

CELL 4: BEST JOB ROLE RECOMMENDATION (AI DECISION)

```python
best_role = max(role_scores, key=role_scores.get)
best_score = role_scores[best_role]

print("\nRecommended Job Role:", best_role)
print("Best ATS Score:", round(best_score*100, 2), "%")
```

CELL 5: Skill Categorization

```python
technical_skills = {"python", "machine", "learning", "nlp", "html", "css", "flask"}
soft_skills = {"communication", "teamwork", "leadership", "problem"}

resume_words = set(resume_text.lower().split())

found_technical = resume_words.intersection(technical_skills)
found_soft = resume_words.intersection(soft_skills)

print("Technical Skills Identified:", found_technical)
print("Soft Skills Identified:", found_soft)
```

CELL 6: Resume Strength Score

```python
resume_strength = best_score * 100

if len(found_technical) >= 4:
    resume_strength += 10
if len(found_soft) >= 2:
    resume_strength += 5

resume_strength = min(resume_strength, 100)
print("Resume Strength Score:", round(resume_strength, 2), "%")
```

CELL 7: AI Resume Improvement Tips

```python
resume_strength = best_score * 100

if len(found_technical) >= 4:
    resume_strength += 10
if len(found_soft) >= 2:
    resume_strength += 5

resume_strength = min(resume_strength, 100)
print("Resume Strength Score:", round(resume_strength, 2), "%")
```

CELL 8: SKILL GAP ANALYSIS

```python
resume_skills = set(resume_text.lower().split())
job_skills = set(job_roles[best_role].lower().split())

missing_skills = job_skills - resume_skills

print("\nMissing Skills for", best_role, ":")
print(missing_skills)
```

CELL 9: AI RESUME FEEDBACK SYSTEM

```python
if best_score < 0.5:
    feedback = "Resume needs improvement. Add more relevant skills."
elif best_score < 0.75:
    feedback = "Resume is good but can be further optimized."
else:
    feedback = "Resume is highly optimized for the job role."

print("\nAI Feedback:", feedback)
```

CELL 10: AI RESUME SUMMARY GENERATOR

```python
def generate_summary(name, role):
    return f"{name} is an aspiring {role} with strong skills in Python and AI, focused on building intelligent and user-friendly applications."

summary = generate_summary(name, best_role)
print("\nGenerated Resume Summary:\n", summary)
```

CELL 11: RESUME PDF GENERATION

```python
file_name = "AI_Enhanced_Resume.pdf"
c = canvas.Canvas(file_name, pagesize=A4)

# Safe fallback in case tips cell was not run
if 'tips' not in globals():
    tips = ["Resume is well optimized based on AI analysis."]

y = 800
c.setFont("Helvetica-Bold", 16)
c.drawString(50, y, name)
y -= 25

c.setFont("Helvetica", 12)
c.drawString(50, y, f"Recommended Role: {best_role}")
y -= 20
c.drawString(50, y, f"ATS Score: {round(best_score*100,2)}%")
y -= 20
c.drawString(50, y, f"Resume Strength: {round(resume_strength,2)}%")
y -= 30

c.setFont("Helvetica-Bold", 14)
c.drawString(50, y, "Profile Summary")
y -= 20
c.setFont("Helvetica", 12)
c.drawString(50, y, summary)
y -= 30

c.setFont("Helvetica-Bold", 14)
c.drawString(50, y, "Technical Skills")
y -= 20
c.setFont("Helvetica", 12)
c.drawString(50, y, ", ".join(found_technical))
y -= 30

c.setFont("Helvetica-Bold", 14)
c.drawString(50, y, "AI Suggestions")
y -= 20
c.setFont("Helvetica", 12)

for tip in tips:
    c.drawString(50, y, f"- {tip}")
    y -= 15

c.save()

print("Enhanced Resume PDF Generated Successfully (No Errors)")
```
